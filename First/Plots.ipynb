```python
ecdc = pd.read_csv("https://opendata.ecdc.europa.eu/covid19/casedistribution/csv", na_values="")
```

```python
from datetime import datetime
ecdc['dateRep'] = ecdc.dateRep.apply(lambda x:datetime.strptime(x, '%d/%m/%Y'))
```

```python
ecdc = ecdc[['dateRep', 'cases', 'countriesAndTerritories', 'geoId']]
```

```python
ecdc = ecdc.sort_values(by='dateRep', ascending=True)
```

```python
# CN, IN, US, DE, NZ, BR
ecdc[ecdc.geoId=='U']
# ecdc[ecdc.countriesAndTerritories=='UAE'].head()
```

```python
def detail(geoId):
    lst = ecdc[ecdc.geoId==geoId].cases.cumsum().tolist()
    new = []
    for i in range(len(lst)):
        if i<7:
            new.append((lst[i]))
        else:
            new.append((lst[i]-lst[i-7]))
    return [lst, new]

lst, new = detail("CN")
plt.plot(lst, new)
plt.show()
```
